```python
%matplotlib notebook

import os
import matplotlib
from data import PartNetDataset
from vis_utils import draw_partnet_objects
from subprocess import call

matplotlib.pyplot.ion()

shapeA = '2369'
shapeB = '40785'
shapeC = '1282'

# compute B-A+C
cmd = 'bash scripts/eval_edittrans_vae_chair_cd.sh %s %s %s' % (shapeA, shapeB, shapeC)
print(cmd)
call(cmd, shell=True)

# results directory
root_dir = '../data/results/exp_vae_chair_cd_edittrans_%s_%s_%s' % (shapeA, shapeB, shapeC)

objA = PartNetDataset.load_object(os.path.join(root_dir, 'shapeA.json'))
objB = PartNetDataset.load_object(os.path.join(root_dir, 'shapeB.json'))
objC = PartNetDataset.load_object(os.path.join(root_dir, 'shapeC.json'))
outD = PartNetDataset.load_object(os.path.join(root_dir, 'output.json'))

objects = [objA, objB, objC, outD]
object_names = ['A', 'B', 'C', 'B-A+C']
    
draw_partnet_objects(objects=objects, 
                     object_names=object_names, 
                     figsize=(10, 5), leafs_only=True, visu_edges=False,
                     sem_colors_filename='../stats/semantics_colors/Chair.txt')
```
